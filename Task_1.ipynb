```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = sns.load_dataset("iris")
```

```python
# Dataset Summary
print("Shape of dataset:", iris.shape)
print("\nColumn names:", iris.columns)
print("\nFirst 5 rows:\n", iris.head())
```

```python

# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(iris["petal_length"], kde=True)
plt.title("Histogram: Petal Length")
plt.xlabel("Petal Length")
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris.drop(columns="species"))
plt.title("Box Plot of Iris Measurements")
plt.xlabel("Features")
plt.ylabel("Values")
plt.show()
```
